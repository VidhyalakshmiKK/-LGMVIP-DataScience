## Image to pencil sketch using python

```python
# importing the required libraries
import cv2
import matplotlib.pyplot as plt
```

```python
# Reading the image
img=cv2.imread("/content/neon-tropical-monstera-leaf-banner.jpg")
```

```python
# Image to be converted to sketch
RGB_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.axis(False)
plt.show()
```

```python
# Converting image to grey scale

grey_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(grey_img,cmap='gray')
plt.axis(False)
plt.show()
```

```python
# Inverting the image

invert_img=cv2.bitwise_not(grey_img)
plt.imshow(invert_img,cmap='gray')
plt.axis(False)
plt.show()
```

```python
# Blurring the image

blur_img=cv2.GaussianBlur(invert_img,(111,111),0)
plt.imshow(blur_img,cmap='gray')
plt.axis(False)
plt.show()
```

```python
# Invert blurred image

invblur_img=cv2.bitwise_not(blur_img)
plt.imshow(invblur_img,cmap='gray')
plt.axis(False)
plt.show()
```

```python
# sketch the image (by performing bit-wise division between the grayscale image and the inverted-blurred image)

sketch_img=cv2.divide(grey_img,invblur_img, scale=220.0)
plt.imshow(sketch_img,cmap='gray')
plt.axis(False)
plt.show()
```

```python
# display the original image Vs the sketch

plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original image', size=18)
plt.imshow(RGB_img)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Sketch', size=18)
rgb_sketch=cv2.cvtColor(sketch_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
